```python
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image, ImageDraw
import numpy as np
import random
import os
from pathlib import Path
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2
import logging

class CelebrityFacePairDataset(Dataset):
    def __init__(self, 
                 root_folder,
                 image_size=256,
                 min_images_per_celebrity=2,
                 mask_coverage_range=(0.2, 0.8),
                 augment_probability=0.7,
                 face_crop_padding=0.3):
        """
        Dataset for celebrity face pair training
        Two inputs: image1 (reference) + masked_image2 â†’ Output: original_image2
        
        Args:
            root_folder (str): Path to folder containing celebrity subfolders
            image_size (int): Output image size (will be square)
            min_images_per_celebrity (int): Minimum images required per celebrity
            mask_coverage_range (tuple): Range of face area to mask (min, max)
            augment_probability (float): Probability of applying augmentations (no noise)
            face_crop_padding (float): Extra padding around face crop (0.3 = 30% padding)
        """
        self.root_folder = Path(root_folder)
        self.image_size = image_size
        self.min_images_per_celebrity = min_images_per_celebrity
        self.mask_coverage_range = mask_coverage_range
        self.augment_probability = augment_probability
        self.face_crop_padding = face_crop_padding
        
        # Setup logging
        logging.basicConfig(level=logging.INFO)
        self.logger = logging.getLogger(__name__)
        
        # Load celebrity data
        self.celebrity_data = self._load_celebrity_data()
        self.valid_celebrities = list(self.celebrity_data.keys())
        
        self.logger.info(f"Loaded {len(self.valid_celebrities)} celebrities with sufficient images")
        
        # Setup transforms
        self.setup_transforms()
        
    def _load_celebrity_data(self):
        """Load and validate celebrity image pairs"""
        celebrity_data = {}
        image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp'}
        
        for celebrity_folder in self.root_folder.iterdir():
            if not celebrity_folder.is_dir():
                continue
                
            celebrity_name = celebrity_folder.name
            image_pairs = []
            
            # Find all valid image-mask pairs
            for image_file in celebrity_folder.iterdir():
                if (image_file.suffix.lower() in image_extensions and 
                    not image_file.name.endswith('_result.png')):
                    
                    mask_file = celebrity_folder / f"{image_file.stem}_result.png"
                    if mask_file.exists():
                        image_pairs.append({
                            'image_path': str(image_file),
                            'mask_path': str(mask_file)
                        })
            
            # Only include celebrities with minimum required images
            if len(image_pairs) >= self.min_images_per_celebrity:
                celebrity_data[celebrity_name] = image_pairs
                
        return celebrity_data
    
    def setup_transforms(self):
        """Setup augmentation transforms (NO noise augmentations)"""
        
        # Clean augmentations for training (no noise/blur/artifacts)
        # Note: These transforms work on both images and masks
        self.augment_transforms = A.Compose([
            A.HorizontalFlip(p=0.5),
            A.Rotate(limit=10, p=0.3),  # Reduced rotation
            A.RandomBrightnessContrast(brightness_limit=0.15, contrast_limit=0.15, p=0.3),
            A.HueSaturationValue(hue_shift_limit=8, sat_shift_limit=12, val_shift_limit=8, p=0.3),
            # Removed: GaussNoise, Blur - no noise/degradation
        ], p=self.augment_probability)
        
        # Final transforms (always applied)
        self.final_transforms = A.Compose([
            A.Resize(self.image_size, self.image_size),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        
        # Mask-specific transforms (no normalization for masks)
        self.mask_transforms = A.Compose([
            A.Resize(self.image_size, self.image_size),
            ToTensorV2()
        ])
    
    def get_face_bounding_box(self, mask):
        """Extract bounding box from segmentation mask"""
        # Convert mask to numpy if it's PIL
        if isinstance(mask, Image.Image):
            mask_np = np.array(mask)
        else:
            mask_np = mask
            
        # Find non-zero pixels (face regions)
        coords = np.argwhere(mask_np > 0)
        if len(coords) == 0:
            # Return full image if no face detected
            h, w = mask_np.shape[:2]
            return [0, 0, w, h]
        
        y_min, x_min = coords.min(axis=0)
        y_max, x_max = coords.max(axis=0)
        
        # Add padding
        h, w = mask_np.shape[:2]
        padding_x = int((x_max - x_min) * self.face_crop_padding)
        padding_y = int((y_max - y_min) * self.face_crop_padding)
        
        x_min = max(0, x_min - padding_x)
        y_min = max(0, y_min - padding_y)
        x_max = min(w, x_max + padding_x)
        y_max = min(h, y_max + padding_y)
        
        return [x_min, y_min, x_max, y_max]
    
    def crop_face_region(self, image, mask):
        """Crop face region from image using mask"""
        bbox = self.get_face_bounding_box(mask)
        x_min, y_min, x_max, y_max = bbox
        
        if isinstance(image, Image.Image):
            cropped_image = image.crop((x_min, y_min, x_max, y_max))
            cropped_mask = mask.crop((x_min, y_min, x_max, y_max))
        else:
            cropped_image = image[y_min:y_max, x_min:x_max]
            cropped_mask = mask[y_min:y_max, x_min:x_max]
            
        return cropped_image, cropped_mask
    
    def create_masked_image(self, image, mask):
        """Create image with partial or full face masking (50% chance of full mask)"""
        # Convert to numpy for processing
        if isinstance(image, Image.Image):
            img_np = np.array(image)
            mask_np = np.array(mask)
        else:
            img_np = image.copy()
            mask_np = mask.copy()
        
        # Find face region
        face_coords = np.argwhere(mask_np > 0)
        if len(face_coords) == 0:
            return Image.fromarray(img_np) if isinstance(image, Image.Image) else img_np
        
        # 50% chance of full face masking, 50% chance of partial masking
        if random.random() < 0.5:
            # FULL FACE MASKING - mask the entire face region
            for y, x in face_coords:
                img_np[y, x] = [0, 0, 0]
            
            self.logger.debug("Applied full face mask")
            
        else:
            # PARTIAL FACE MASKING - use existing logic
            # Calculate how much of the face to mask
            coverage = random.uniform(*self.mask_coverage_range)
            n_pixels_to_mask = int(len(face_coords) * coverage)
            
            # Randomly select pixels to mask
            mask_indices = random.sample(range(len(face_coords)), n_pixels_to_mask)
            pixels_to_mask = face_coords[mask_indices]
            
            # Create different clean masking patterns
            mask_type = random.choice(['black_patches', 'partial_face', 'random_regions'])
            
            if mask_type == 'black_patches':
                # Create clean black rectangular patches
                y_coords, x_coords = pixels_to_mask[:, 0], pixels_to_mask[:, 1]
                y_min, y_max = y_coords.min(), y_coords.max()
                x_min, x_max = x_coords.min(), x_coords.max()
                
                # Create 1-3 clean rectangular patches
                n_patches = random.randint(1, 3)
                for _ in range(n_patches):
                    patch_h = random.randint(15, (y_max - y_min) // 2)
                    patch_w = random.randint(15, (x_max - x_min) // 2)
                    start_y = random.randint(y_min, max(y_min + 1, y_max - patch_h))
                    start_x = random.randint(x_min, max(x_min + 1, x_max - patch_w))
                    
                    img_np[start_y:start_y + patch_h, start_x:start_x + patch_w] = [0, 0, 0]
            
            elif mask_type == 'partial_face':
                # Mask specific face parts cleanly
                y_coords, x_coords = face_coords[:, 0], face_coords[:, 1]
                y_center = (y_coords.min() + y_coords.max()) // 2
                x_center = (x_coords.min() + x_coords.max()) // 2
                
                mask_region = random.choice(['top', 'bottom', 'left', 'right', 'center'])
                
                if mask_region == 'top':
                    mask_pixels = face_coords[face_coords[:, 0] < y_center]
                elif mask_region == 'bottom':
                    mask_pixels = face_coords[face_coords[:, 0] > y_center]
                elif mask_region == 'left':
                    mask_pixels = face_coords[face_coords[:, 1] < x_center]
                elif mask_region == 'right':
                    mask_pixels = face_coords[face_coords[:, 1] > x_center]
                else:  # center
                    # Mask central region
                    y_quarter = (y_coords.max() - y_coords.min()) // 4
                    x_quarter = (x_coords.max() - x_coords.min()) // 4
                    mask_pixels = face_coords[
                        (face_coords[:, 0] > y_center - y_quarter) & 
                        (face_coords[:, 0] < y_center + y_quarter) &
                        (face_coords[:, 1] > x_center - x_quarter) & 
                        (face_coords[:, 1] < x_center + x_quarter)
                    ]
                
                for y, x in mask_pixels:
                    img_np[y, x] = [0, 0, 0]
                    
            else:  # random_regions
                # Clean random black masking
                for y, x in pixels_to_mask:
                    img_np[y, x] = [0, 0, 0]
            
            self.logger.debug(f"Applied partial face mask: {mask_type}, coverage: {coverage:.2f}")
        
        return Image.fromarray(img_np) if isinstance(image, Image.Image) else img_np
    
    def __len__(self):
        # Each celebrity can generate multiple pairs
        total_pairs = sum(len(images) * (len(images) - 1) for images in self.celebrity_data.values())
        return total_pairs // 2  # Divide by 2 to avoid duplicate pairs
    
    def __getitem__(self, idx):
        # Randomly select a celebrity
        celebrity = random.choice(self.valid_celebrities)
        celebrity_images = self.celebrity_data[celebrity]
        
        # Randomly select two different images from the same celebrity
        if len(celebrity_images) < 2:
            # Fallback to different celebrity (shouldn't happen due to filtering)
            celebrity = random.choice(self.valid_celebrities)
            celebrity_images = self.celebrity_data[celebrity]
        
        image1_data, image2_data = random.sample(celebrity_images, 2)
        
        try:
            # Load images and masks
            image1 = Image.open(image1_data['image_path']).convert('RGB')
            mask1 = Image.open(image1_data['mask_path']).convert('L')
            image2 = Image.open(image2_data['image_path']).convert('RGB')
            mask2 = Image.open(image2_data['mask_path']).convert('L')
            
            # Crop face regions
            face1, face_mask1 = self.crop_face_region(image1, mask1)
            face2, face_mask2 = self.crop_face_region(image2, mask2)
            
            # INPUT 1: Reference image (image1) - always complete
            input1 = face1
            
            # OUTPUT: Original image2 (complete, unmasked)
            target = face2
            
            # Apply augmentations first, then create mask
            if random.random() < self.augment_probability:
                # Generate seeds for consistent augmentation
                seed1 = random.randint(0, 999999)
                seed2 = random.randint(0, 999999)
                
                # Augment input1 (reference) with its own seed
                random.seed(seed1)
                np.random.seed(seed1)
                augmented_input1 = self.augment_transforms(image=np.array(input1))['image']
                input1 = Image.fromarray(augmented_input1)
                
                # Augment image2 and mask2 with the same seed (BEFORE masking)
                random.seed(seed2)
                np.random.seed(seed2)
                augmented_target = self.augment_transforms(image=np.array(target))['image']
                target = Image.fromarray(augmented_target)
                
                # Augment mask2 with the same seed as image2
                random.seed(seed2)
                np.random.seed(seed2)
                augmented_mask2 = self.augment_transforms(image=np.array(face_mask2))['image']
                face_mask2 = Image.fromarray(augmented_mask2)
            
            # NOW create masked version using the augmented image2 and mask2
            masked_image2 = self.create_masked_image(target, face_mask2)
            
            # Apply final transforms
            input1_tensor = self.final_transforms(image=np.array(input1))['image']
            input2_tensor = self.final_transforms(image=np.array(masked_image2))['image']
            target_tensor = self.final_transforms(image=np.array(target))['image']
            
            # Convert masks to tensors (for potential loss computation)
            mask1_tensor = self.mask_transforms(image=np.array(face_mask1))['image']
            mask2_tensor = self.mask_transforms(image=np.array(face_mask2))['image']
            
            return {
                'input1': input1_tensor,        # Reference image (complete face)
                'input2': input2_tensor,        # Masked image2
                'target': target_tensor,        # Ground truth (original image2)
                'mask1': mask1_tensor,          # Segmentation mask for input1
                'mask2': mask2_tensor,          # Segmentation mask for target/input2
                'celebrity': celebrity          # Celebrity name for debugging
            }
            
        except Exception as e:
            self.logger.error(f"Error loading data for {celebrity}: {e}")
            # Return a different random sample
            return self.__getitem__(random.randint(0, len(self) - 1))

def create_dataloader(root_folder, 
                     batch_size=16, 
                     num_workers=4, 
                     shuffle=True, 
                     **dataset_kwargs):
    """
    Create DataLoader for the celebrity face dataset
    
    Args:
        root_folder (str): Path to celebrity folders
        batch_size (int): Batch size for training
        num_workers (int): Number of worker processes
        shuffle (bool): Whether to shuffle data
        **dataset_kwargs: Additional arguments for dataset
    
    Returns:
        DataLoader: Ready-to-use dataloader
    """
    dataset = CelebrityFacePairDataset(root_folder, **dataset_kwargs)
    
    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        num_workers=num_workers,
        pin_memory=torch.cuda.is_available(),
        drop_last=True
    )
    
    return dataloader

# ============ USAGE EXAMPLE ============

if __name__ == "__main__":
    # Example usage
    root_folder = "human_Dataset/Original Images"
    
    # Create dataset
    dataset = CelebrityFacePairDataset(
        root_folder=root_folder,
        image_size=256,
        min_images_per_celebrity=3,
        mask_coverage_range=(0.2, 0.7),
        augment_probability=0.8
    )
    
    print(f"Dataset size: {len(dataset)}")
    print(f"Available celebrities: {len(dataset.valid_celebrities)}")
    
    # Create dataloader
    dataloader = create_dataloader(
        root_folder=root_folder,
        batch_size=8,
        num_workers=2,
        image_size=256
    )
    
    # Test loading a batch
    for batch_idx, batch in enumerate(dataloader):
        print(f"Batch {batch_idx}:")
        print(f"  Input1 (reference) shape: {batch['input1'].shape}")
        print(f"  Input2 (masked) shape: {batch['input2'].shape}")
        print(f"  Target (original) shape: {batch['target'].shape}")
        print(f"  Mask1 shape: {batch['mask1'].shape}")
        print(f"  Mask2 shape: {batch['mask2'].shape}")
        print(f"  Celebrities: {batch['celebrity']}")
        
        if batch_idx == 0:  # Just show first batch
            break

# ============ VISUALIZATION HELPER ============

def visualize_dataset_batch(dataloader, num_samples=4):
    """Visualize samples from the dataset"""
    import matplotlib.pyplot as plt
    
    batch = next(iter(dataloader))
    
    fig, axes = plt.subplots(3, num_samples, figsize=(15, 10))
    
    # Denormalize images for visualization
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
    
    for i in range(num_samples):
        # Denormalize
        input1_img = batch['input1'][i] * std + mean
        input2_img = batch['input2'][i] * std + mean
        target_img = batch['target'][i] * std + mean
        
        # Convert to PIL for display
        input1_img = transforms.ToPILImage()(torch.clamp(input1_img, 0, 1))
        input2_img = transforms.ToPILImage()(torch.clamp(input2_img, 0, 1))
        target_img = transforms.ToPILImage()(torch.clamp(target_img, 0, 1))
        
        # Check if input2 is fully masked (mostly black)
        input2_np = np.array(input2_img)
        black_ratio = np.sum(input2_np.sum(axis=2) < 30) / (input2_np.shape[0] * input2_np.shape[1])
        mask_type = "FULL MASK" if black_ratio > 0.7 else "PARTIAL MASK"
        
        # Display
        axes[0, i].imshow(input1_img)
        axes[0, i].set_title(f'Input1: Reference\n{batch["celebrity"][i]}')
        axes[0, i].axis('off')
        
        axes[1, i].imshow(input2_img)
        axes[1, i].set_title(f'Input2: {mask_type}')
        axes[1, i].axis('off')
        
        axes[2, i].imshow(target_img)
        axes[2, i].set_title(f'Target: Original')
        axes[2, i].axis('off')
    
    plt.tight_layout()
    plt.show()

# ============ TESTING MASK DISTRIBUTION ============

def test_mask_distribution(dataloader, num_samples=100):
    """Test the distribution of full vs partial masks"""
    full_mask_count = 0
    partial_mask_count = 0
    
    print("Testing mask distribution...")
    
    for batch_idx, batch in enumerate(dataloader):
        batch_size = batch['input2'].shape[0]
        
        # Denormalize images for analysis
        mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
        std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
        
        for i in range(batch_size):
            input2_img = batch['input2'][i] * std + mean
            input2_img = transforms.ToPILImage()(torch.clamp(input2_img, 0, 1))
            
            # Check if mostly black (full mask)
            input2_np = np.array(input2_img)
            black_ratio = np.sum(input2_np.sum(axis=2) < 30) / (input2_np.shape[0] * input2_np.shape[1])
            
            if black_ratio > 0.7:  # More than 70% black = full mask
                full_mask_count += 1
            else:
                partial_mask_count += 1
        
        # Stop after num_samples
        if (batch_idx + 1) * batch_size >= num_samples:
            break
    
    total = full_mask_count + partial_mask_count
    print(f"\nMask Distribution Analysis ({total} samples):")
    print(f"Full masks: {full_mask_count} ({full_mask_count/total*100:.1f}%)")
    print(f"Partial masks: {partial_mask_count} ({partial_mask_count/total*100:.1f}%)")
    print(f"Target: 50% each")
    
    return full_mask_count, partial_mask_count

def example_training_loop():
    """Example of how to use this dataset in training"""
    
    # Create dataloader
    dataloader = create_dataloader(
        root_folder="human_Dataset/Original Images",
        batch_size=16,
        image_size=256,
        mask_coverage_range=(0.3, 0.6)
    )
    
    # Example training loop
    for epoch in range(10):
        for batch_idx, batch in enumerate(dataloader):
            # Extract inputs and target
            input1 = batch['input1']  # Reference image [B, 3, 256, 256]
            input2 = batch['input2']  # Masked image [B, 3, 256, 256]
            target = batch['target']  # Original image [B, 3, 256, 256]
            
            # Concatenate inputs or feed separately to your model
            # Option 1: Concatenate inputs
            model_input = torch.cat([input1, input2], dim=1)  # [B, 6, 256, 256]
            
            # Option 2: Feed separately (if your model takes two inputs)
            # output = model(input1, input2)
            
            # Your model forward pass here
            # output = model(model_input)
            # loss = criterion(output, target)
            
            print(f"Epoch {epoch}, Batch {batch_idx}: "
                  f"Input1: {input1.shape}, Input2: {input2.shape}, Target: {target.shape}")
            
            if batch_idx == 2:  # Just for demo
                break
```
