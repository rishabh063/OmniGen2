```python
from datasets import load_dataset

# Load dataset
dataset = load_dataset('Yuanshi/Subjects200K')
```

```python
def filter_func(item):
    if item.get("collection") != "collection_2":
        return False
    if not item.get("quality_assessment"):
        return False
    return all(
        item["quality_assessment"].get(key, 0) >= 5
        for key in ["compositeStructure", "objectConsistency", "imageQuality"]
    )

dataset = dataset["train"].filter(
    filter_func,
    num_proc=16,
    cache_file_name="./cache/dataset/collection_2_valid.arrow", # Optional
)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from PIL import Image

model = AutoModelForCausalLM.from_pretrained(
    "vikhyatk/moondream2",
    revision="2025-06-21",
    trust_remote_code=True,
    device_map={"": "cuda"}  # ...or 'mps', on Apple Silicon
)


```

```python
def scalebbox(imgsize , bboxes):
    img_width, img_height = imgsize
    boxarray=[]
    for bbox in bboxes:
        x_min = bbox['x_min']
        y_min = bbox['y_min']
        x_max = bbox['x_max']
        y_max = bbox['y_max']
        
        # Convert normalized coordinates to pixel coordinates if needed
        # (assuming coordinates are normalized if they're all <= 1)
        # if x_min <= 1 and y_min <= 1 and x_max <= 1 and y_max <= 1:
        x_min = int(x_min * img_width)
        y_min = int(y_min * img_height)
        x_max = int(x_max * img_width)
        y_max = int(y_max * img_height)
        boxarray.append([x_min ,y_min , x_max ,y_max  ])  
    return boxarray

from PIL import Image, ImageDraw

def draw_bboxes(image, bboxes, box_color='red', line_width=2):
    """
    Draw bounding boxes on a PIL image.
    
    Args:
        image: PIL Image object
        bboxes: List of bounding boxes in format [(x1, y1, x2, y2), ...] or single bbox (x1, y1, x2, y2)
        box_color: Color of the bounding box lines (default: 'red')
        line_width: Width of the bounding box lines (default: 2)
    
    Returns:
        PIL Image with bounding boxes drawn
    """
    # Create a copy to avoid modifying the original image
    img_copy = image.copy()
    draw = ImageDraw.Draw(img_copy)
    
    # Handle single bbox or list of bboxes
    if isinstance(bboxes[0], (int, float)):
        # Single bbox case: (x1, y1, x2, y2)
        bboxes = [bboxes]
    
    # Draw each bounding box
    for bbox in bboxes:
        x1, y1, x2, y2 = bbox
        # Draw rectangle outline
        draw.rectangle([x1, y1, x2, y2], outline=box_color, width=line_width)
    
    return img_copy

# Example usage:
# image = Image.open('your_image.jpg')
# bbox = (50, 50, 200, 150)  # Single bbox
# # or multiple bboxes: [(50, 50, 200, 150), (100, 100, 250, 200)]
# result = draw_bboxes(image, bbox)
# result.show()  # Display the image
# result.save('output.jpg')  # Save the image
```

```python
def black_out_bboxes(image, bboxes):
    # Create a copy to avoid modifying the original image
    img_copy = image.copy()
    draw = ImageDraw.Draw(img_copy)
    
    # Handle single bbox or list of bboxes
    if isinstance(bboxes[0], (int, float)):
        # Single bbox case: (x1, y1, x2, y2)
        bboxes = [bboxes]
    
    # Draw black filled rectangles over each bounding box
    for bbox in bboxes:
        x1, y1, x2, y2 = bbox
        # Draw filled black rectangle
        draw.rectangle([x1, y1, x2, y2], fill='black')
    
    return img_copy
def bigbbox(img,bboxes):
    min_x,min_y=img.size
    max_x,max_y=0,0
    for bbox in bboxes:
        if bbox[0]<min_x:
            min_x=bbox[0]
        if bbox[1]<min_y:
            min_y=bbox[1]
        if bbox[2]>max_x:
            max_x=bbox[2]
        if bbox[3]>max_y:
            max_y=bbox[3]
    return [min_x,min_y,max_x,max_y]
def addPadding(img,bboxes,padding):
    for idx in range(0,len(bboxes)):
        min_x,min_y,max_x,max_y=bboxes[idx]
        min_x=max(0,min_x-padding)
        min_y=max(0,min_y-padding)
        max_x=min(img.size[0],max_x+padding)
        max_y=min(img.size[1],max_y+padding)
        bboxes[idx]=[min_x,min_y,max_x,max_y]
    return bboxes
def create_images(img,bboxes,padding=10):
    
    if random.random()<0.5:
        big_bbox=bigbbox(img,bboxes)
        print("without padding")
        ref_img=img.crop(big_bbox)
    else:
        print("with padding")
        big_bbox=bigbbox(img,addPadding(img,bboxes,padding))
        ref_img=img.crop(big_bbox)

    if random.random()<0.5:
        print("without padding")
        mask_image= black_out_bboxes(img,bboxes)
    else:
        print("with padding")
        tempbboxes=addPadding(img,bboxes,padding)  
        mask_image=black_out_bboxes(img,tempbboxes)
    return ref_img , mask_image
```
