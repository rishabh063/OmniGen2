```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import random

def view_garments(folder_name=None, max_folders=10, randomize=True):
    """
    View original garments, masks, and clipped garments side by side with image sizes
    
    Args:
        folder_name: Specific folder to view (if None, shows random folders)
        max_folders: Maximum number of folders to display
        randomize: If True, shows random folders each time
    """
    simple_data_path = "simple_data2"
    
    if folder_name:
        folders = [folder_name] if os.path.exists(os.path.join(simple_data_path, folder_name)) else []
    else:
        all_folders = [f for f in os.listdir(simple_data_path) 
                      if os.path.isdir(os.path.join(simple_data_path, f))]
        
        if randomize and len(all_folders) > max_folders:
            folders = random.sample(all_folders, max_folders)
            print(f"Showing {max_folders} random folders out of {len(all_folders)} total folders")
        else:
            folders = all_folders[:max_folders]
    
    if not folders:
        print("No folders found or specified folder doesn't exist")
        return
    
    # Calculate grid size
    n_folders = len(folders)
    fig, axes = plt.subplots(n_folders, 3, figsize=(15, 5*n_folders))
    
    # Handle single folder case
    if n_folders == 1:
        axes = axes.reshape(1, -1)
    
    for i, folder in enumerate(folders):
        folder_path = os.path.join(simple_data_path, folder)
        
        # File paths
        garment_path = os.path.join(folder_path, "garment.jpg")
        mask_path = os.path.join(folder_path, "garment_mask.png")
        clipped_path = os.path.join(folder_path, "clipped_garment.png")
        
        # Plot original garment
        if os.path.exists(garment_path):
            img = Image.open(garment_path)
            axes[i, 0].imshow(img)
            size_text = f"{img.size[0]}{img.size[1]}"
            axes[i, 0].set_title(f"{folder}\nOriginal Garment\nSize: {size_text}")
            axes[i, 0].axis('off')
        else:
            axes[i, 0].text(0.5, 0.5, 'No garment.jpg', ha='center', va='center')
            axes[i, 0].set_title(f"{folder}\nOriginal Garment\nSize: N/A")
            axes[i, 0].axis('off')
        
        # Plot mask
        if os.path.exists(mask_path):
            mask = Image.open(mask_path)
            axes[i, 1].imshow(mask, cmap='gray')
            size_text = f"{mask.size[0]}{mask.size[1]}"
            axes[i, 1].set_title(f"Garment Mask\nSize: {size_text}")
            axes[i, 1].axis('off')
        else:
            axes[i, 1].text(0.5, 0.5, 'No mask found', ha='center', va='center')
            axes[i, 1].set_title("Garment Mask\nSize: N/A")
            axes[i, 1].axis('off')
        
        # Plot clipped garment
        if os.path.exists(clipped_path):
            clipped = Image.open(clipped_path)
            axes[i, 2].imshow(clipped)
            size_text = f"{clipped.size[0]}{clipped.size[1]}"
            axes[i, 2].set_title(f"Clipped Garment\nSize: {size_text}")
            axes[i, 2].axis('off')
        else:
            axes[i, 2].text(0.5, 0.5, 'No clipped image', ha='center', va='center')
            axes[i, 2].set_title("Clipped Garment\nSize: N/A")
            axes[i, 2].axis('off')
    
    plt.tight_layout()
    plt.show()

def get_image_size_summary(folder_name=None):
    """
    Get a summary of image sizes across all folders
    
    Args:
        folder_name: Specific folder to analyze (if None, analyzes all folders)
    """
    simple_data_path = "simple_data2"
    
    if folder_name:
        folders = [folder_name] if os.path.exists(os.path.join(simple_data_path, folder_name)) else []
    else:
        folders = [f for f in os.listdir(simple_data_path) 
                  if os.path.isdir(os.path.join(simple_data_path, f))]
    
    if not folders:
        print("No folders found")
        return
    
    sizes = {'garment': [], 'mask': [], 'clipped': []}
    
    for folder in folders:
        folder_path = os.path.join(simple_data_path, folder)
        
        # Check each image type
        for img_type, filename in [('garment', 'garment.jpg'), 
                                  ('mask', 'garment_mask.png'), 
                                  ('clipped', 'clipped_garment.png')]:
            img_path = os.path.join(folder_path, filename)
            if os.path.exists(img_path):
                try:
                    with Image.open(img_path) as img:
                        sizes[img_type].append(img.size)
                except Exception as e:
                    print(f"Error reading {img_path}: {e}")
    
    # Print summary
    print(f"\nImage Size Summary ({len(folders)} folders analyzed):")
    print("-" * 50)
    
    for img_type, size_list in sizes.items():
        if size_list:
            unique_sizes = list(set(size_list))
            print(f"\n{img_type.capitalize()} images:")
            print(f"  Total count: {len(size_list)}")
            print(f"  Unique sizes: {len(unique_sizes)}")
            for size in unique_sizes:
                count = size_list.count(size)
                percentage = (count / len(size_list)) * 100
                print(f"    {size[0]}{size[1]}: {count} images ({percentage:.1f}%)")
        else:
            print(f"\n{img_type.capitalize()} images: No images found")

# Usage examples:
# View random folders with sizes
# view_garments()

# View specific folder with sizes
# view_garments("your_folder_name")

# View more random folders at once with sizes
view_garments(max_folders=20)

# Get size summary for all images
# get_image_size_summary()

# Get size summary for specific folder
# get_image_size_summary("your_folder_name")
```
